```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
#Import the Dataset
df = pd.read_excel('/content/drive/MyDrive/Github/m_tech/Datasets/input.xlsx')
df.rename(columns={'SalesVolume':'Values'},inplace=True)
```

```python
len(df)
```

```python
df['weighted'] = df['Values'].ewm(span = 20, adjust = False).mean()
```

```python
plt.figure(figsize = (12,6))
plt.plot(df['weighted'], label="Values")
# plt.plot(df['weighted'], label="weighted")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```
